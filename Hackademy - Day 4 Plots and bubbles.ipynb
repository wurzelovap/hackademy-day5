```python
# libraries
import matplotlib.pyplot as plt
import pandas

# create data
data = pandas.read_csv("/home/start/Desktop/data.csv", index_col = "year")
data_2002 = data.loc[2002]
GDP_2002 = data_2002["gdpPercap"]
LifeExp_2002 = data_2002["lifeExp"]
Population_2002 = data_2002["pop"]/1e6
Continent_2002 = data_2002["continent"]
country = data_2002["country"]


def colorify(Continent_2002):
    colours = []
    for country in Continent_2002:
        if country == "Asia":
            colours.append("red")
        elif country == "Europe":
            colours.append("green")
        elif country == "Africa":
            colours.append("blue")
        elif country == "Americas":
            colours.append("black")
        elif country == "Oceania":
            colours.append("yellow")
    return colours


x = GDP_2002
y = LifeExp_2002
z = Population_2002

# Change color with c and alpha. I map the color to the X axis value.
plt.scatter(x, y, z, c=colorify(Continent_2002), cmap="Blues", alpha=0.6, edgecolors="grey", linewidth=2)

# Add titles (main and on axis)
plt.xlabel("GDP Per Capita")
plt.ylabel("Life Expectancy")
plt.title("Wenzel's Fantastic Bubble Plot")

plt.show()
```
